```python
file = "dictionary.csv"
countries_df = pd.read_csv(file, decimal=",")
countries_df.head(15)
```

```python
file = "summer.csv"
summer_df = pd.read_csv(file, decimal=",")
summer_df.head(15)
```

```python
file = "winter.csv"
winter_df = pd.read_csv(file, decimal=",")
winter_df.head(15)
```

```python
summer_df = summer_df.rename(columns={"Country":"Code"})
```

```python
summer_countries_df = summer_df.merge(countries_df, on="Code")
summer_countries_df["Season"] = "Summer" 
summer_countries_df.head(15)
```

```python
winter_df = winter_df.rename(columns={"Country":"Code"})
winter_countries_df = winter_df.merge(countries_df, on="Code")
winter_countries_df["Season"] = "Winter" 
winter_countries_df.head(15)
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df], sort = False)
all_df
```

```python
all_df.groupby(["Country"]).count()["Medal"]
```

```python
all_count_df = all_df[all_df["Year"] >= 1984].groupby(["Country"]).count()[["Medal"]].sort_values(by="Medal", ascending=False)
all_count_df.head(10)
```
